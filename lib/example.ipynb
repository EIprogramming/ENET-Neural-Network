```python
%load_ext autoreload
%reload_ext autoreload
%autoreload 2
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from neural_net import NeuralNet
import mushrooms as shroom
import graphics
```

```python
from sklearn.model_selection import train_test_split

X, y = shroom.generate_mushrooms(lambda i, j: (i-0.5)**2 + (j-0.5)**2 < 0.16, size_shrooms=5000)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

```python
epochs = 350
neural_net = NeuralNet((2, 16, 1), learning_rate=0.001, random_state=42, dtype=np.float32)
losses, y_pred = neural_net.train(X_train, y_train, epochs=epochs, validate=(X_test, y_test), batch_size=6, binary=True)
```

```python
shroom.plot_mushrooms(X, y)
y_pred = neural_net.predict(X_test)
shroom.plot_mushrooms(X_test, y_pred)
graphics.plot_losses(epochs, losses)
```
